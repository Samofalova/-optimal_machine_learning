```python
from sklearn.datasets import make_blobs
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_score, recall_score, accuracy_score
import pandas as pd
import time
```

# 1

```python
X, y = make_blobs(n_samples=10000, centers=2, n_features=2, random_state=1, cluster_std=3)
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 3)
```

```python
start_time = time.time()
test1_1 = own_logistic_regression(x_train, y_train, x_test, draw = True)
end_time = time.time()
```

```python
pr1_1 = precision_score(y_test, test1_1['y_hat'])
rc1_1 = recall_score(y_test, test1_1['y_hat'])
ac1_1 = accuracy_score(y_test, test1_1['y_hat'])
tm1_1 = end_time - start_time
print(f'Precision = {pr1_1}, recall = {rc1_1}, accuracy = {ac1_1}, time = {tm1_1}')
```

```python
start_time = time.time()
test2_1 = logistic_regression_with_rbf(x_train, y_train, x_test, draw=True)
end_time = time.time()
```

```python
pr2_1 = precision_score(y_test, test2_1['y_hat'])
rc2_1 = recall_score(y_test, test2_1['y_hat'])
ac2_1 = accuracy_score(y_test, test2_1['y_hat'])
tm2_1 = end_time - start_time
print(f'Precision = {pr2_1}, recall = {rc2_1}, accuracy = {ac2_1}, time = {tm2_1}')
```

```python
start_time = time.time()
test3_1 = svm_with_pddipm(x_train, y_train, x_test, draw=True)
end_time = time.time()
```

```python
pr3_1 = precision_score(y_test, test3_1['y_hat'])
rc3_1 = recall_score(y_test, test3_1['y_hat'])
ac3_1 = accuracy_score(y_test, test3_1['y_hat'])
tm3_1 = end_time - start_time
print(f'Precision = {pr3_1}, recall = {rc3_1}, accuracy = {ac3_1}, time = {tm3_1}')
```

```python
start_time = time.time()
test4_1 = Ð¡lassification_linear_regression_l1(x_train, y_train, x_test, draw = True)
end_time = time.time()
test4_1
```

```python
pr4_1 = precision_score(y_test, test4_1['y_pred'])
rc4_1 = recall_score(y_test, test4_1['y_pred'])
ac4_1 = accuracy_score(y_test, test4_1['y_pred'])
tm4_1 = end_time - start_time
print(f'Precision = {pr4_1}, recall = {rc4_1}, accuracy = {ac4_1}, time = {tm4_1}')
```
