```python
# ТЗ 3

def f(x, y):
    return 4*x**2 + 4*x + y**2 - 8*y+5

def f_diff(x, y):
    return np.array([8*x + 4, 2*y - 8])

def f_diff_2x(x, y):
    return np.array([8, 2])

def f_diff_xy(x, y):
    return np.array([0, 0])

%time
res = newton_algorithm(f, f_diff, f_diff_2x, f_diff_xy)
```

3. $0.26(x^2 + y^2) - 0.48xy$

```python
# ТЗ 1

%time
main()
```

```python
# ТЗ 3

def f(x, y):
    return 0.26 * (x ** 2 + y ** 2) - 0.48 * x * y

def f_diff(x, y):
    return np.array([0.52 * x - 0.48 * y,
                     -0.48 * x + 0.52 * y])
def f_diff_2x(x, y):
    return np.array([0.52, 0.52])

def f_diff_xy(x, y):
    return np.array([-0.48,-0.48])

%time
res = newton_algorithm(f, f_diff, f_diff_2x, f_diff_xy)
```

```python
res # Считаем, что нули
```

```python
# Т3 6

func = '0.26 * (x ** 2 + y ** 2) - 0.48 * x * y'
restrictions = []

%time 
y = log_barriers(func, restrictions, (0, 0))
```

```python
y
```
