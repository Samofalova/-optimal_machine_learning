# Генерация с помощью make_blobs, 3 признака


```python
X, y = make_blobs(n_samples = 400, centers=2, cluster_std=1.2, n_features=3, random_state=42)
for i,j in enumerate(y):
    if j == 0:
        y[i] = -1
    elif j == 1:
        y[i] = 1

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
X_train.shape, y_train.shape
```

```python
own_pegasos(X_train, y_train, X_test, draw = False)
```

```python
y_test
```
